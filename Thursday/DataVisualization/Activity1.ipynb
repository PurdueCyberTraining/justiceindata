# Data Visualization 
1. We will make different plots using Matplotlib and Seaborn libraries. 
2. The dataset contains daily temperature data of two different places for one year.  

```python
# This will download file Data_Vis1.xlsx from Github 
import requests 

csv_xlsx = "https://raw.githubusercontent.com/stm5131/nsfcybertraining/main/Data_Vis/Data_Vis1.xlsx" # donwloading raw url

response = requests.get(csv_xlsx) 

if response.status_code == 200: 

    with open('Data_Vis1.xlsx', 'wb') as f: 
        f.write(response.content) 
        print("File downloaded successfully.") 
else: 
    print('Failed to download file.') 
```

```python
#calling required libraries
import pandas as pd
import matplotlib.pyplot as plt  
import seaborn as sns
```

```python
#Read file, load this file 
df = pd.read_excel("Data_Vis1.xlsx")
```

```python
#reading the dataframe
df
```

# Line plots using Matplotlib

```python
#assigning names to data in file
Place1 = df ['Place1']
Place2  = df ['Place2']
Date = df ['Day']
#plt.plot used for plotting line graphs
plt.plot(Date, Place1, label='Place1')
plt.plot(Date, Place2, label='Place2')
#labelling x and y axis
plt.xlabel('Day')
plt.ylabel('Temp(°C)')
#plot title
plt.title('Temperature variation')
#plot legend
plt.legend(loc='upper left')
#saving image. It will be saved in the current working directory
plt.savefig('1'+ '.jpeg',
               format='jpeg',
                 dpi=600)
# for showing plot
plt.show()
```

```python
# changing line color 
#plt.plot used for plotting line graphs , You can use codes: r for red, g for green, k for black, w for white etc.
plt.plot(Date, Place1, label='Place1', color='r')
plt.plot(Date, Place2, label='Place2', color='g')
#labelling x and y axis
plt.xlabel('Day')
plt.ylabel('Temp(°C)')
#plot title
plt.title('Temperature variation')
#plot legend
plt.legend(loc='upper left')
```

```python
# changing line widths
#plt.plot used for plotting line graphs
plt.plot(Date, Place1, label='Place1', linewidth=4)
plt.plot(Date, Place2, label='Place2', linewidth=0.5)
#labelling x and y axis
plt.xlabel('Day')
plt.ylabel('Temp(°C)')
#plot title
plt.title('Temperature variation')
#plot legend
plt.legend(loc='upper left')
```

```python
# changing line style
#plt.plot used for plotting line graphs 
plt.plot(Date, Place1, label='Place1', linestyle='solid')
plt.plot(Date, Place2, label='Place2', linestyle='dashed')
#labelling x and y axis
plt.xlabel('Day')
plt.ylabel('Temp(°C)')
#plot title
plt.title('Temperature variation')
#plot legend
plt.legend(loc='upper left')
```

```python
# changing font size
#plt.plot used for plotting line graphs
plt.plot(Date, Place1, label='Place1', linestyle='solid')
plt.plot(Date, Place2, label='Place2', linestyle='dashed')
#labelling x and y axis
plt.xlabel('Day',fontsize=15)
plt.ylabel('Temp(°C)',fontsize=15)
#plot title
plt.title('Temperature variation',fontsize=15)
#plot legend
plt.legend(loc='upper left',fontsize=15)
# font size of both axis
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
```

```python
# changing axis limits
#plt.plot used for plotting line graphs
plt.plot(Date, Place1, label='Place1', linestyle='solid')
plt.plot(Date, Place2, label='Place2', linestyle='dashed')
#labelling x and y axis
plt.xlabel('Day',fontsize=15)
plt.ylabel('Temp(°C)',fontsize=15)
#plot title
plt.title('Temperature variation',fontsize=15)
#plot legend
plt.legend(loc='upper left',fontsize=15)
# font size of both axis
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
# setting axis limit
plt.xlim([25, 50])
plt.ylim([10, 30])
```

# Line plots using Seaborn

```python
sns.lineplot(x = "Day", y = "Place1", data=df) 
sns.lineplot(x = "Day", y = "Place2", data=df)
plt.xlabel('Day')
plt.ylabel('Temp(°C)')
plt.legend(labels=["Place1","Place2"])
plt.title('Temperature variation')
plt.show()
```

```python
#Hue function
sns.lineplot(x = "Day", y = "Place2", data=df, hue='Month',palette="dark") 
plt.xlabel('Day')
plt.ylabel('Temp(°C)')
plt.legend(loc='upper left')
plt.title('Temperature variation')
plt.show()
```

# Changing color palettes 
You can find different color palettes from https://seaborn.pydata.org/tutorial/color_palettes.html .Let's change color palette to 'deep' and 'rocket'. 

```python
#Hue function with 'deep' palette
sns.lineplot(x = "Day", y = "Place2", data=df, hue='Month',palette="deep") 
plt.xlabel('Day')
plt.ylabel('Temp(°C)')
plt.legend(loc='upper left')
plt.title('Temperature variation')
plt.show()
```

```python
#Hue function with a wrong palette
sns.lineplot(x = "Day", y = "Place2", data=df, hue='Month',palette="paired") # there is no palette called paired. However, in error we can see all available palettes
plt.xlabel('Day')
plt.ylabel('Temp(°C)')
plt.legend(loc='upper left')
plt.title('Temperature variation')
plt.show()
```

```python
#Hue function with a chosen palette
sns.lineplot(x = "Day", y = "Place2", data=df, hue='Month',palette="hot") 
plt.legend(loc='upper left')
plt.xlabel('Day')
plt.ylabel('Temp(°C)')
plt.title('Temperature variation')
```

# Bar graph using Matplotlib

```python
#stack chart 
plt.bar(Date, Place1, label='Place1')
plt.bar(Date, Place2, label='Place2',bottom=Place1)

#labelling x and y axis
plt.xlabel('Day')
plt.ylabel('Temp(°C)')
#plot title
plt.title('Temperature variation')
#plot legend
plt.legend(loc='upper left')
# for showing plot
plt.show()
```

# Violin plots

```python
#using seaborn library
# for Place1 
sns.violinplot( y=Place1 );
plt.xlabel('Location-Place1')
plt.ylabel('Temp(°C)')
plt.show()
```

```python
# making violin plots for each month for Place1
sns.violinplot( x=df["Month"], y=df["Place1"], palette="Blues")
plt.xlabel('Months')
plt.ylabel('Temp(°C)')
plt.show()
```

```python
# making violin plots for each month for Place1 and removing inner box plots to clearly see violin plots
sns.violinplot( x=df["Month"], y=df["Place1"], palette="Blues", inner=None)
plt.xlabel('Months')
plt.ylabel('Temp(°C)')
plt.show()
```

```python
#making violin plot for Place1 and Place2
df.drop(['Day', 'Month'], axis=1, inplace=True) # if we don't drop 'Day' and 'Month column', sns.boxplot will create boxplots for these as well which is not required.
sns.violinplot(data=df)
plt.xlabel('Location')
plt.ylabel('Temp(°C)')
plt.show()
```
