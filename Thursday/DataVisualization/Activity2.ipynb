# Creating multiple plots 
    The file 'Multiple Places' contains daily rainfall (mm) for 8 different places.

```python
# This will download file Multiple_places.xlsx from Github 
import requests 

csv_xlsx = "https://raw.githubusercontent.com/stm5131/nsfcybertraining/main/Data_Vis/Multiple_places.xlsx" # donwloading raw url

response = requests.get(csv_xlsx) 

if response.status_code == 200: 

    with open('Multiple_places.xlsx', 'wb') as f: 
        f.write(response.content) 
        print("File downloaded successfully.") 
else: 
    print('Failed to download file.') 
```

```python
#calling required libraries
import pandas as pd
import matplotlib.pyplot as plt  
import seaborn as sns
```

```python
#Read file, load this file 
df = pd.read_excel("Multiple_places.xlsx")
```

```python
df
```

```python
# rainfall variation of all places in one image
#assigning names to data in file
for i in range(8): # we want to iterate 8 times because we have 8 places
    y = df.iloc[:,i+1] # this one locates the columnn 
    x = df ['Day'] # as on x-axis, we want column named 'Day', we spcecify its name because it is not iterable and fixed
    #plt.plot used for plotting line graphs
    plt.plot(x, y, label='Place'+str(i+1))
    #labelling x and y axis
    plt.xlabel('Day')
    plt.ylabel('Rainfall (mm)')
    #plot title
    plt.title('Rainfall variation')
    #plot legend
    plt.legend(fontsize="8", loc='upper left')
plt.savefig('AllPlaces'+ '.jpeg',
               format='jpeg',
                 dpi=600)
plt.show()
```

```python
#Changing legend location
for i in range(8): # we want to iterate 8 times because we have 8 places
    y = df.iloc[:,i+1] # this one locates the columnn 
    x = df ['Day'] # as on x-axis, we want column named 'Day', we spcecify its name because it is not iterable and fixed
    #plt.plot used for plotting line graphs
    plt.plot(x, y, label='Place'+str(i+1))
    #labelling x and y axis
    plt.xlabel('Day')
    plt.ylabel('Rainfall (mm)')
    #plot title
    plt.title('Rainfall variation')
    #plot legend
    plt.legend(ncol=5,fontsize="8") # displays legend in 5 columns
plt.show()
```

```python
# plot for each place
for i in range(8):
    y = df.iloc[:,i+1]
    x = df ['Day']
    #plt.plot used for plotting line graphs
    plt.plot(x, y, label='Place'+str(i+1))
    #labelling x and y axis
    plt.xlabel('Day')
    plt.ylabel('Rainfall (mm)')
    #plot title
    plt.title('Rainfall variation')
    #plot legend
    plt.legend(loc='upper left')
    #plt.savefig('Place' +str(i+1)+ '.jpeg',
              # format='jpeg',
                # dpi=600)
    plt.show()
    plt.clf() # it is required to clear the plot otherwise the next plot will be plotted on the earlier one
```
