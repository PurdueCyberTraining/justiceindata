# Data-Preprocessing #2 :
    This activity will be performed using a file having discharge measurements from a USGS station.

```python
# This will download file Measurements.csv from Github 
import requests 

csv_url = "https://raw.githubusercontent.com/stm5131/nsfcybertraining/main/Data/Measurements.csv" # donwloading raw url

response = requests.get(csv_url) 

if response.status_code == 200: 

    with open('Measurements.csv', 'wb') as f: 
        f.write(response.content) 
        print("File downloaded successfully.") 
else: 
    print('Failed to download file.') 
```

```python
# import required library 
import pandas as pd
```

```python
#load csv file
df = pd.read_csv('Measurements.csv')
df
```

# Requirements
1. Dishcarge measurments with 'Good'measured rating diff
2. Discharge measurements with party name 'LPH'
3. Dicharge measurments meeting above mentioned both criteria 

```python
# 1.Dishcarge measurments with 'Good'measured rating diff

rslt_df = df.loc[(df['measured_rating_diff'] == 'Good')]
rslt_df.reset_index()
```

```python
# 2. Discharge measurements with party name 'LPH'

options = ['LPH']
rslt_df1 = df.loc[df['party_nm'].isin(options)]
rslt_df1.reset_index()
```

```python
# 3. Discharge measurements with party name 'LPH' and 'Good'measured rating diff

options = ['Good']
rslt_df2 = df.loc[(df['party_nm']=='LPH')&df['measured_rating_diff'].isin(options)]
rslt_df2.reset_index()
```

# Plotting discharge with measurements from LPH with good rating diff

```python
import matplotlib.pyplot as plt
x = rslt_df2.measurement_nu
y = rslt_df2.discharge_va
plt.plot(x,y, label='Discharge variation with measurements')
plt.xlabel('Measurement_number')
plt.ylabel('Discharge (cfs)')
plt.legend(loc='upper left')
plt.show()
```
