But what if we want only the primary education buildings and we do not need the others. We will need to pre-process the unneccesary data out.


Now let's look at buildings that are all in the same class. PrimClass is Primary Education buildings so we will look at those first.

```python
primary_buildings = df.timestamp #Create a new empty space for the data we want.

for column in df.columns: #Look over every column in our raw data
    
    if column.split("_")[0]=="PrimClass": #Check if the beginning of this column name is what we want.
        
        primary_buildings = pd.concat(
            [primary_buildings, df[column]],axis=1) #If it is then add it to the new dataframe
        
        
```

This loop looks at every column inside the data frame and checks if it starts with PrimClass.
It does this by splitting the string into a list of strings split by the "_" character.

If it is a primary building class then it adds it to a new data frame we are using.

Notice how the date and time are just a string and the computer does not know how to interpret them? We can adjust this.

```python
primary_buildings['timestamp'] = pd.to_datetime(primary_buildings['timestamp'])
```
