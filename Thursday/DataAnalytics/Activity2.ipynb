# Finding coefficient of determination (R²)

```python
# This will download file temp1.csv from Github 
import requests 

csv_url = "https://raw.githubusercontent.com/stm5131/nsfcybertraining/main/Data/temp_1.csv" # donwloading raw url

response = requests.get(csv_url) 

if response.status_code == 200: 

    with open('temp_1.csv', 'wb') as f: 
        f.write(response.content) 
        print("File downloaded successfully.") 
else: 
    print('Failed to download file.') 
```

```python
# This will download file temp2.csv from Github 
csv_url = "https://raw.githubusercontent.com/stm5131/nsfcybertraining/main/Data/temp_2.csv" # donwloading raw url

response = requests.get(csv_url) 

if response.status_code == 200: 

    with open('temp_2.csv', 'wb') as f: 
        f.write(response.content) 
        print("File downloaded successfully.") 
else: 
    print('Failed to download file.') 
```

```python
#Import libraries
import pandas as pd
from sklearn.metrics import r2_score 
```

```python
#Read file
df1 = pd.read_csv("temp_1.csv")
df2 = pd.read_csv("temp_2.csv")
```

```python
# See dataframe 1 having temp_1.csv file 
df1
```

```python
# See dataframe 2 having temp_2.csv file 
df2
```

```python
# Renaming columns using indexing
df1.columns.values[0] = "Date"
df1.columns.values[1] = "Place1"
#df1
```

```python
# Renaming columns using indexing
df2.columns.values[0] = "Date"
df2.columns.values[1] = "Place2"
#df2
```

```python
#Combining both dataframes
Combine= pd.concat([df1,df2], axis=1)
```

```python
Combine
```

```python
#Drop columns. We are doing it to see only temperature values for both places next to each other in the dataframe
Combine.drop(['Date'], axis=1) # if we use inplace=True, it wil make actual changes in the dataframe
```

```python
# Finding coefficient of determination value using r2_score function of scikit-learn library
a = Combine.Place1
b = Combine.Place2

R_square = r2_score(a, b) 
```

```python
R_square
```

```python
#Save dataframe (combine) to a csv file
Combine.to_csv('Combine_Temp'+'.csv', index= False) #index 'False' means we don't need indexing in the saved csv files
```
